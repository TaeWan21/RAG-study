```python
import json
from tqdm import tqdm
import openai
```

```python
user_prompt = df['content'].loc[15]
```

```python
user_prompt
```

```python
import os
from dotenv import load_dotenv
load_dotenv()
key = os.getenv("OPENAI_API_KEY")
client = openai.OpenAI(api_key= key)
```

```python
def get_analysys_from_news(user_prompt):
  response = client.chat.completions.create(
    model="gpt-4o",
    messages=[
          {"role": "system", "content": system_prompt},
          {"role": "user", "content": user_prompt + '\n답변:'}
      ],
    temperature=0
  )
  return response.choices[0].message.content
```

```python
analysis_result = get_analysys_from_news(user_prompt)
print(analysis_result)
```

```python
# 1000개만 사용
user_inputs = df['content'].to_list()[:1000]
```

```python
result_lst = []
for user_prompt in tqdm(user_inputs):
  result_lst.append(get_analysys_from_news(user_prompt))
```

```python
df = df[:1000]
```

```python
df['result'] = result_lst
```

```python
# True/False 변환 및 JSON 파싱
def parse_json(row):
    try:
        # True/False를 소문자로 변환 후 JSON 로드
        corrected = row.replace("True", "true").replace("False", "false")
        return json.loads(corrected)
    except json.JSONDecodeError:
        return None  # 파싱 실패 시 None 반환

# JSON 파싱 적용
df['parsed_result'] = df['result'].apply(parse_json)

# 파싱 실패한 행 제거
df_cleaned = df.dropna(subset=['parsed_result']).reset_index(drop=True)
```

```python
# 결과 확인
df_cleaned
```

```python
df_cleaned['system_prompt'] = system_prompt
df_cleaned['user_prompt'] = df_cleaned['content']
df_cleaned['assistant'] = df_cleaned['parsed_result']

df_cleaned[['system_prompt', 'user_prompt', 'assistant']]
```

```python
output_file = "train_data.json"

df_cleaned[['system_prompt', 'user_prompt', 'assistant']].to_json(
    output_file, orient="records", force_ascii=False, indent=4
)
```
